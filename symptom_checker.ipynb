```python
import pandas as pd

df = pd.read_csv('Symptomchecker.csv') 
symptoms_by_condition = {}

unique_health_conditions = df['prognosis'].unique()
for condition in unique_health_conditions:
    
    condition_df = df[df['prognosis'] == condition]
    
    condition_symptoms = condition_df.columns[:-1]  
    condition_symptoms = condition_symptoms[condition_df.iloc[0, :-1] == 1]  
    
    
    symptoms_by_condition[condition] = list(condition_symptoms)


for condition, symptoms in symptoms_by_condition.items():
    print(f'Condition: {condition}')
    print(f'Symptoms: {", ".join(symptoms)}')
    print()
```

```python
import json

with open('data.json', 'r') as json_file:
    medical_data = json.load(json_file)

def predict_condition(symptoms):
    matching_conditions = []
    
    for condition, data in medical_data.items():
        condition_symptoms = data['Symptoms']
        matching_symptoms = set(symptoms) & set(condition_symptoms)
        
        if len(matching_symptoms) >= 4:
            matching_conditions.append(condition)
    
    return matching_conditions

input_symptoms = [
    "fatigue",
    "weight_loss",
    "restlessness",
    "lethargy",
    "irregular_sugar_level",
    "blurred_and_distorted_vision",
    "obesity",
    "excessive_hunger",
    "increased_appetite",
    "polyuria"
]

predicted_conditions = predict_condition(input_symptoms)

if predicted_conditions:
    print("Predicted Condition(s):")
    for condition in predicted_conditions:
        print(condition)
else:
    print("No matching condition found.")
```
